```python
import numpy as np
def decimal_to_binary(decimal):
    binary_recov = ""
    while decimal!= 0:
        decimal *= 2
        bit = int(decimal)
        decimal -= bit
        binary_recov += str(bit)
    return binary_recov

def binary_to_decimal(binary_trunc):
    decimal = 0
    for i in range(len(binary_trunc)):
        bit = int(binary_trunc[i])
        decimal += 2**(-(i+1))*bit
    return decimal

alphabet = ['a','b','c','d','e']
pmf = np.array([0.5,0.25,0.125,0.0625,0.0625])
cdf = np.insert(np.cumsum(pmf),0,0)
cdf = cdf[:-1]
print("cdf: ", cdf)
pmf_map = {}
cdf_map = {}
for i in range(len(alphabet)):
    pmf_map[alphabet[i]] = pmf[i]
    cdf_map[alphabet[i]] = pmf[:i].sum()
stream_len = 500
for seed in range(100):
    np.random.seed(seed)
    stream = []
    for s,p in pmf_map.items():
        stream += [s]*int(p*stream_len)
    np.random.shuffle(stream)

    stream = "".join(stream)
    # stream = "bbb"
    stream_len = len(stream)

    lower_bound = 0
    upper_bound = 1
    lower_real = 0
    upper_real = 1
    length_real = 1
    encoded = ""
    for s in stream:
        lower_real = lower_real + length_real * cdf_map[s]
        upper_real = lower_real + length_real * pmf_map[s]
        length_real *= pmf_map[s]
        length = upper_bound-lower_bound
        lower_bound = lower_bound + length * cdf_map[s]
        upper_bound = lower_bound + length * pmf_map[s]
        while int(lower_bound*2)==int(upper_bound*2):
            encoded += str(int(lower_bound*2))
            lower_bound = lower_bound*2 - int(lower_bound*2)
            upper_bound = upper_bound*2 - int(upper_bound*2)
        length = upper_bound-lower_bound
        # print(f"{s},{lower_bound:20},{upper_bound:20}, {length:.3f}, {encoded}")
    decoded = ""
    decode_lower = 0
    decode_upper = 1
    shift = 0
    length = 1
    interval_double = 0
    i = 0
    while len(decoded)<len(stream):
        if i<len(encoded):
            b = int(encoded[i])
        else: b = 0 if i==len(encoded) else 1
        decode_lower += b*2**(shift-i-1)
        decode_upper = decode_lower + 2**(shift-i-1)
        interval_double += 1
        flter = np.logical_and(cdf>=decode_lower, cdf<decode_upper)
        if flter.sum()==1:
            # extract symbol
            s = alphabet[np.argmax(flter)]
            decoded += s
            # update interval
            decode_lower = (decode_lower-cdf_map[s])/pmf_map[s]
            decode_upper = decode_lower + pmf_map[s]
            shift = i+1
        i += 1
    print(decoded==stream)
    if not decoded==stream:
        print(decoded)
        print(stream)
```
